## Prepare
* Import Modules
* Load the dataset
* Create a tokenizer and fit it on the text data.
* Get the number of unique words

```python
import pandas as pd
from keras.preprocessing.text import Tokenizer

df = pd.read_csv('train.csv')
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['text'].values)

num_unique_words = len(tokenizer.word_index)
```

* import modules
* Create the feature matrix and target vector
* Split the data into training and testing sets

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential, Model
from keras.layers import LSTM, Dense, Embedding, Dropout, Input
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import SpatialDropout1D, Bidirectional
from tensorflow.keras.regularizers import L2
from keras import regularizers


max_features = num_unique_words
maxlen = 33
tokenizer = Tokenizer(num_words=max_features, split=' ')
tokenizer.fit_on_texts(df['text'].values)
X = tokenizer.texts_to_sequences(df['text'].values)
X = pad_sequences(X)

y = df['target']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Build and Compile a LSTM Model
### What's LSTM?
The central role of an LSTM model is held by a memory cell known as a ‘cell state’ that maintains its state over time. The cell state is the horizontal line that runs through the top of the below diagram. It can be visualized as a conveyor belt through which information just flows, unchanged.

```python
import tensorflow as tf

inputs = Input(shape=(maxlen,))
x = Embedding(max_features, 128, input_length=maxlen)(inputs)
x = LSTM(128, return_sequences=True)(x)
x = Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.01))(x)
x = Dropout(0.2)(x)
x = Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01))(x)
x = Dropout(0.2)(x)
x = LSTM(64)(x)
x = Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01))(x)
x = Dropout(0.2)(x)
outputs = Dense(1, activation='sigmoid')(x)

model = Model(inputs=inputs, outputs=outputs)


optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)
model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
model.summary()
```

## Train the model

```python
!pip install livelossplot
from livelossplot.tf_keras import PlotLossesCallback
batch_size = 32
epochs = 100
model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_test, y_test)
          ,callbacks=[PlotLossesCallback()])
```

## Predict on the test set AND Save the prediction to a file

```python
test_df = pd.read_csv('test.csv')
tokenizer.fit_on_texts(test_df['text'].values)
X = tokenizer.texts_to_sequences(test_df['text'].values)
X = pad_sequences(X)

y_pred = model.predict(X)

results_df = pd.DataFrame({'id': test_df['id'], 'target': y_pred.reshape(-1)})
results_df.to_csv('results.csv', index=False)
```
