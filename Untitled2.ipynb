```python
import tensorflow as tf
import tensorflow.keras as keras
import matplotlib.pyplot as plt
import numpy as np
import zipfile
from IPython import display
```

```python
def plot_results(images, n_cols=None):
  display.clear_output(wait=False)
  n_cols = n_cols or len(images)
  n_rows = (len(images) - 1) // n_cols + 1
  if images.shape[-1] == 1:
    images = np.squeeze(images, axis=-1)
  plt.figure(figsize=(n_cols, n_rows))
  for index, image in enumerate(images):
    plt.subplot(n_rows, n_cols, index + 1)
    plt.imshow(image, cmap="binary")
    plt.axis("off")
```

```python
training_file_name = "/content/photo_jpg.zip"
training_dir = "/content/train"
zip_ref = zipfile.ZipFile(training_file_name, 'r')
zip_ref.extractall(training_dir)
zip_ref.close()
```

```python
BATCH_SIZE = 32

def map_images(file):
  img = tf.io.decode_png(tf.io.read_file(file))
  img = tf.dtypes.cast(img, tf.float32)
  img = img / 255.0
  return img
```

```python
filename_dataset = tf.data.Dataset.list_files("/content/train/*.jpg")
image_dataset = filename_dataset.map(map_images).batch(BATCH_SIZE) 
```

```python
random_normal_dimensions = 256
```

```python
generator = keras.models.Sequential([
  keras.layers.Dense(8*8*256,input_shape=[random_normal_dimensions]),
  keras.layers.Reshape([8,8,256]),
  keras.layers.BatchNormalization(),
  keras.layers.Conv2DTranspose(256,6,4,padding='same',activation='relu'),
  keras.layers.BatchNormalization(),
  keras.layers.Conv2DTranspose(3,7,8,padding='same',activation='tanh')
])
```

```python
generator.summary()
```

```python
discriminator = keras.models.Sequential([
  keras.layers.Conv2D(64, kernel_size=5, strides=2, padding="SAME", activation=keras.layers.LeakyReLU(0.2), input_shape=[256,256,3]),
  keras.layers.Dropout(0.4),
  keras.layers.Conv2D(128, kernel_size=5, strides=2, padding="SAME", activation=keras.layers.LeakyReLU(0.2)),
  keras.layers.Dropout(0.4),
  keras.layers.Flatten(),
  keras.layers.Dense(1, activation="sigmoid")
])
```

```python
discriminator.summary()
```

```python
discriminator.compile(loss='binary_crossentropy', optimizer='rmsprop')
discriminator.trainable = False
```

```python
gan = keras.models.Sequential([generator, discriminator])
gan.compile(loss='binary_crossentropy', optimizer='rmsprop')
```

```python
def train_gan(gan, dataset, random_normal_dimensions, n_epochs=50):
  generator, discriminator = gan.layers
  for epoch in range(n_epochs):
    print("Epoch {}/{}".format(epoch + 1, n_epochs))
    for real_images in dataset:
      real_batch_size = real_images.shape[0]
      noise = tf.random.normal(shape=[real_batch_size, random_normal_dimensions])
      fake_images = generator(noise)
      mixed_images = tf.concat([fake_images, real_images], axis=0)
      discriminator_labels = tf.constant([[0.]] * real_batch_size + [[1.]] * real_batch_size)
      discriminator.trainable = True
      discriminator.train_on_batch(mixed_images, discriminator_labels)
      noise = tf.random.normal([real_batch_size, random_normal_dimensions])
      generator_labels = tf.constant([[1.]] * real_batch_size)
      discriminator.trainable = False
      gan.train_on_batch(noise, generator_labels)
    plot_results(fake_images, 16)
    plt.show()
  return fake_images
```

```python
EPOCHS = 50
fake_images = train_gan(gan, image_dataset, random_normal_dimensions, EPOCHS)
```

```python
from PIL import Image
from zipfile import ZipFile

denormalized_images = fake_images * 255
denormalized_images = tf.dtypes.cast(denormalized_images, dtype = tf.uint8)
file_paths = []
```

```python
for this_image in range(len(fake_images)):
  #print(denormalized_images[this_image].shape)
  i = tf.reshape(denormalized_images[this_image], [256,256,3])
  im = Image.fromarray(i.numpy())
  im = im.convert("L")
  filename = "hand" + str(this_image) + ".jpg"
  file_paths.append(filename)
  im.save(filename)
```

```python
with ZipFile('my-signs.zip', 'w') as zip:
  for file in file_paths:
    zip.write(file)
```
